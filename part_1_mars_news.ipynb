````python

import requests
from bs4 import BeautifulSoup

# URL of the Mars news website (replace with actual URL)
url = 'https://static.bc-edx.com/data/web/mars_news/index.html'  

# Step 1: Send a GET request to fetch the page content
response = requests.get(url)

# Step 2: Parse the content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Extract specific elements (e.g., headlines, links, summaries)
# Example of extracting article headlines
headlines = soup.find_all('h2', class_='headline')  # Adjust tag/class based on the actual site structure

# Loop through the headlines and print the text content
for headline in headlines:
    print(headline.get_text(strip=True))

# Example of extracting links within headlines
links = soup.find_all('a', href=True)
for link in links:
    print(f"Link: {link['href']} - Text: {link.get_text(strip=True)}")

# Example of extracting summaries (if they exist)
summaries = soup.find_all('p', class_='summary')  # Adjust tag/class based on actual site structure
for summary in summaries:
    print(summary.get_text(strip=True))

# If you want to print the entire HTML (for inspection purposes)
# print(soup.prettify())



````

````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Initialize WebDriver
options = webdriver.ChromeOptions()
options.headless = True  # Run in headless mode (no UI)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Step 1: Load the page
url = 'https://static.bc-edx.com/data/web/mars_news/index.html'  # Replace with actual URL
driver.get(url)

# Give the page some time to load dynamic content
time.sleep(5)  # Adjust sleep time as needed

# Step 2: Get the page source after JavaScript has loaded
html = driver.page_source

# Step 3: Parse the content with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Step 4: Extract titles and preview text
articles = soup.find_all('div', class_='news-article')  # Adjust this to the correct tag/class

article_data = []
for article in articles:
    # Adjust these selectors based on actual HTML structure
    title_tag = article.find('h2', class_='headline')  # Adjust the tag and class name
    preview_tag = article.find('p', class_='preview-text')  # Adjust the tag and class name

    if title_tag and preview_tag:
        title = title_tag.get_text(strip=True)
        preview = preview_tag.get_text(strip=True)
        article_data.append({'title': title, 'preview': preview})

# Step 5: Beautiful Loop for Output
print("\n--- Mars News Articles ---\n")
for idx, item in enumerate(article_data, start=1):
    print(f"Article {idx}:")
    print(f"  Title   : {item['title']}")
    print(f"  Preview : {item['preview']}")
    print("-" * 40)  # Adds a separator line

# Close the browser
driver.quit()
# After the soup is created, print the prettified HTML
print(soup.prettify())  # Add this to inspect the HTML content
````
